```python
from collections import  Counter 
popular_interests  = Counter(interest for user_interests in users_interests
                                       for interest in user_interests)
```

```python
popular_interests
```

```python
from typing import List,Tuple
def most_popular_new_interest(
    user_interests: List[str],
    max_results:int =5) -> List[Tuple[str,int]]:
    suggestins = [(interest,frequency) for interest,frequency in popular_interests.most_common()
                                         if interest not in user_interests]
    return suggestins[:max_results]

    

most_popular_new_interest(users_interests[0])
```

```python
## 以下代码中{} 表示Set,自动去重

unique_interests = sorted({interest for user_interests in  users_interests 
                                    for interest in user_interests} )
unique_interests
```

```python
def make_user_interest_vector(user_interests: List[str]) -> List[int]:
    return [1  if interest in user_interests else 0    for interest in unique_interests]

make_user_interest_vector(users_interests[0])
```

```python
user_interest_vectors = [make_user_interest_vector(user_interests) for user_interests in users_interests]
user_interest_vectors
```

```python
import numpy as np
user_interest_vectors = np.array(user_interest_vectors)

```

```python


def consine_similarity(x,y, eps = 1e-8):
    nx = x / (np.sqrt(np.sum(x**2)) + eps)
    ny = y / (np.sqrt(np.sum(y**2)) + eps)
    return np.dot(nx,ny)
    
    #$nv = v /np.sqrt(np.sum(v ** 2))
   # return np.dot(v,w) / np.sqrt(np.dot(v,v),np.dot(w,w))




    
    
user_interest_matrix = np.array(list(map(make_user_interest_vector,users_interests)))



## 余弦相似度，参考深度学习进阶：自然语言处理






user_similarities = [[consine_similarity(interest_vector_i,interest_vector_j) for interest_vector_j in user_interest_matrix] 
                   for interest_vector_i in user_interest_matrix]



n = user_interest_matrix/(np.sqrt(np.sum(user_interest_matrix ** 2,axis=1,keepdims=True))+1e-8)
user_similarities_2 = np.dot(n, n.T)



```

```python
def most_similar_user_to(user_id):

    pairs = [(other_user_id,similarity) for other_user_id,similarity in enumerate(user_similarities[0])
        if user_id != other_user_id and similarity > 0]
    return sorted(pairs,key=lambda pair: pair[0],reverse=0)


def most_similar_user_to2(user_id):

    pairs = [(other_user_id,similarity) for other_user_id,similarity in enumerate(user_similarities_2[0])
        if user_id != other_user_id and similarity > 0]
    return sorted(pairs,key=lambda pair: pair[0],reverse=0)


print(most_similar_user_to(3))

print(most_similar_user_to2(3))
#most_similar_user_to(1)

```
