```python
import re
import torch
import numpy as np
```

```python
file=open("/Users/wzp/Workspace/learn/pytorch_code/04/cls_reg/housing.data",mode='r')

data =[re.split(r"\s{1,}",item.strip()) for item in file]
data = np.array(data).astype(np.float)
   
```

```python
y ,x = data[:,-1],data[:,0:-1]

print(x.shape)
print(y.shape)
```

```python
x_train , y_train   = x[0:496,...],  y[0:496,...]
x_test  , y_test    = x[496:,...],    y[496:,...]
```

```python
class Net(torch.nn.Module):
    def __init__(self,n_feature,n_output):
        super(Net,self).__init__()
        self.hidden=torch.nn.Linear(n_feature,100)
        self.predict=torch.nn.Linear(100,n_output)
    def forward(self,x):
        out = self.hidden(x)
        out = torch.relu(out)
        out = self.predict(out)
        return out
    
        
```

```python
net = Net(13,1)
loss_func = torch.nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(),lr=0.001)
```

```python
x_data ,y_data = torch.tensor(x_train,dtype=torch.float32), torch.tensor(y_train,dtype=torch.float32)
print(x_data,y_data)
```

```python
for i in range(30000):
    pred = net.forward(x_data)
    torch.squeeze(pred)
    loss = loss_func(pred,y_data)*0.001
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print("iter:{},loss:{}".format(i,loss))
    
```

```python
x_data ,y_data = torch.tensor(x_test,dtype=torch.float32), torch.tensor(y_test,dtype=torch.float32)
print(x_data,y_data)
```

```python
pred = net.forward(x_data)
pred = torch.squeeze(pred)
loss_test = loss_func(pred, y_data) * 0.001
print("ite:{}, loss_test:{}".format(i, loss_test))
```
